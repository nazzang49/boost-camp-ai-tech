```python
import csv
```

```python
# test => 입력된 명령어 빈도 확인
command_data = []
with open("command_data.csv", "r", encoding='utf-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in spamreader:
        command_data.append(row)
command_data
```

```python
# 빈도 확인용 Dict
command_count = dict()
for value in command_data:
    if value[1] in command_count.keys():
        command_count[value[1]] += 1
    else:
        command_count[value[1]] = 1
command_count
```

```python
# Dict to List
count_list = list()
for key, value in command_count.items():
    count_list.append([key, value])
count_list
```

```python
# 내림차순 정렬
def getKey(item):
    return item[1]

# 정렬 대상 리스트 / 정렬 기준값 / 내림차순 여부
ans = sorted(count_list, key = getKey, reverse=True)
ans[:10]
```
