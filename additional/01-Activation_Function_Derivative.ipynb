## tanh

```python
'''
    Args:
        x: matrix
        return: original / derivative
'''
def tanh(x):
    t = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    dt = 1 - t ** 2
    return t, dt
```

```python
t, dt = tanh(0)
print(dt)
```
